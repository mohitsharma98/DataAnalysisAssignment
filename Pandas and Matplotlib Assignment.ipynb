```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
df = pd.read_csv("data.csv", low_memory=False)
```

```python
df['report_date'] = pd.to_datetime(df['report_date'], errors='coerce')
df.drop(['time_period','time_period_type'],axis=1, inplace=True)
```

```python
new1 = df[df['unit']=='cases']
new1.drop(['unit','data_field','data_field_code','location_type'], axis=1, inplace=True)
new1.set_index('location', inplace=True)
new1['value'] = pd.to_numeric(new1['value'], errors='coerce')
```

# Third Question

```python
dnew = df[['report_date','location', 'data_field', 'value']]
dnew['value'] = pd.to_numeric(dnew['value'], errors='coerce')
dnew.set_index('location', inplace=True)
```

```python
dnewnew = dnew.groupby(by=['location','data_field']).sum()
buenos_val = dnewnew.loc['Argentina-Buenos_Aires']['value']
caba_val = dnewnew.loc['Argentina-CABA']['value']
cata_val = dnewnew.loc['Argentina-Catamarca']['value']
x_axis = dnew.loc['Argentina-Buenos_Aires']['data_field'].unique()
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,6))
slices = [buenos_val, caba_val, cata_val]
p = ['Argentina-Buenos_Aires','Argentina-CABA','Argentina-Catamarca']

for i in range(3):
    axes[i].title.set_text(p[i])
    axes[i].pie(slices[i], startangle=90, wedgeprops={"edgecolor":"0",'linewidth': 0.5,
                    'linestyle': 'solid', 'antialiased': True})
    axes[i].axis('equal')
    plt.legend(loc='lower left' ,labels=x_axis, bbox_to_anchor=(1, 0.8))
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(13,9))
plt.plot(x_axis, cata_val, lw=3, ls='--', label='Argentina-CABA')
plt.plot(x_axis, caba_val, lw=3, ls='solid', label='Argentina-Catamarca')
plt.plot(x_axis, buenos_val, lw=3, ls='-.', label='Argentina-Buenos_Aires')
plt.xlabel('Data Field Values')
plt.xticks(rotation=45)
plt.ylabel('Values')
plt.legend()
plt.tight_layout()
```

```python
data = df[['location', 'data_field', 'value']]
data['value'] = pd.to_numeric(data.value, errors='coerce')
```

```python
yes = data.query('location == "Argentina-Buenos_Aires" or location == "Argentina-CABA" or location == "Argentina-Catamarca"').groupby(by=['data_field', 'location'], as_index=False).sum()
yes.set_index('data_field', inplace=True)
dval = list(yes.index.unique())
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
cityList = ['Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca']
colors = ['lightskyblue','gold', 'lightcoral']
p=0
for i in range(2):
    for j in range(3):
        l = list(yes.loc[dval[p]]['value'])
        ax[i,j].pie(l, colors=colors, autopct='%1.0d%%', explode=(0,0.1,0), wedgeprops={"edgecolor":"0",'linewidth': 1,
                    'linestyle': 'solid', 'antialiased': True})
        ax[i,j].title.set_text(dval[p])
        p+=1
plt.title('Distribution of Data Field Values According to Cities')
plt.legend(loc='lower left', labels=cityList, fontsize='x-large')
plt.tight_layout(pad=3)
```

```python
from difflib import get_close_matches
import re
```

```python
def words(text):
    return re.findall(r'\w+', text.lower())

wordSet = set(words(open('big.txt').read()))
```

```python
print('Type "exit" to end the Spell Checker.')
while(True):
    str1 = input('Enter a Word:')
    correction = get_close_matches(str1, wordSet, n=2,cutoff=0.5)
    if str1=='exit' or str1=='Exit':
        break
    elif correction[0] == str1:
        print('Your word is correct!')
    else:
        print('Did you mean: {}?'.format(correction))
```
